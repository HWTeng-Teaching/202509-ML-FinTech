(i) In this data set, how many of the suburbs average more than seven rooms per dwelling? More than eight rooms per dwelling? Comment on the suburbs that average more than eight rooms per dwelling.

```python
count_7 = Boston[Boston["rm"] > 7].shape[0]
count_8 = Boston[Boston["rm"] > 8].shape[0]

print(f"There are {count_7} suburbs with average rooms > 7, "
      f"and {count_8} suburbs with average rooms > 8.")
```

**Commemt**: The group which the suburbs average more than 8 rooms per dwelling features <font color=#00000FF>lower percentage of the status of the population</font>. And for the unsupervised learning field, it might form a clear and perfact cluster.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假設 Boston 已經讀入
# Boston = pd.read_csv("Boston.csv")

# 建立 Over8 欄位
Boston["Over8"] = (Boston["rm"] > 8).astype(int)

# 取得除了 rm 與 Over8 的所有變數
other_cols = Boston.columns.drop(["rm", "Over8"])

# 設定顏色
colors = {0: "blue", 1: "red"}

# 設定子圖：4列 x 3行
fig, axes = plt.subplots(4, 3, figsize=(18, 20))
axes = axes.flatten()  # 方便用一維 index 存取

for i, col in enumerate(other_cols):
    ax = axes[i]
    for over8_value, group in Boston.groupby("Over8"):
        ax.scatter(group["rm"], group[col],
                   color=colors[over8_value], alpha=0.7, label=f"Over8={over8_value}")
    ax.set_xlabel("rm")
    ax.set_ylabel(col)
    ax.set_title(f"rm vs {col}")
    ax.legend()

# 如果子圖數量比變數多，關掉多餘的子圖
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
