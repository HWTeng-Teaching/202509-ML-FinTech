````python
cols = ["Outstate", "PhD", "Grad.Rate","S.F.Ratio","Top10perc","Room.Board"]

# 1. 直方圖
college[cols].hist(bins=30, figsize=(10, 6))
plt.suptitle("Distributions of Full-time, Part-time Undergraduates and Graduation Rate")
plt.tight_layout()
plt.show()
````

````python
import seaborn as sns
# 2. 兩兩散點圖 (pairplot)
sns.pairplot(college[cols], diag_kind="hist")
plt.suptitle("Pairwise Scatter Plots", y=1.02)
plt.show()
````

````python
# 3. 相關係數矩陣
corr = college[cols].corr()
print("Correlation matrix:\n", corr)
````

````python
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
````
