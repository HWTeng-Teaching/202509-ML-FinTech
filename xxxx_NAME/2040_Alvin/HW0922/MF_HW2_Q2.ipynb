## Problem 2

Suppose that we have four observations, for which we compute a dissimilarity matrix, given by
$$\begin{pmatrix} & 0.3 & 0.4 & 0.7\\ 0.3 & & 0.5 & 0.8 \\0.4 & 0.5 & & 0.45 \\0.7 & 0.8 & 0.45 \end{pmatrix}$$
For instance, the dissimilarity between the first and second observations is 0.3, and the dissimilarity between the second and fourth
observations is 0.8.

(a) On the basis of this dissimilarity matrix, sketch the dendrogram that results from hierarchically clustering these four observations using complete linkage. Be sure to indicate on the plot the height at which each fusion occurs, as well as the observations corresponding to each leaf in the dendrogram.

We first import the packages and do some preparation framework.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from ISLP.cluster import compute_linkage
from scipy.spatial.distance import squareform
```

```python
# The dissimilarity matrix provided in the problem
dissimilarity_matrix = np.array([
    [0.0, 0.3, 0.4, 0.7],
    [0.3, 0.0, 0.5, 0.8],
    [0.4, 0.5, 0.0, 0.45],
    [0.7, 0.8, 0.45, 0.0]
])

# Perform hierarchical clustering using 'complete' linkage.
# The `linkage` function from `scipy.cluster.hierarchy` takes a condensed
# distance matrix. We can convert our square matrix to condensed format.
# A condensed matrix is a 1D array of the upper triangular entries.
condensed_dissimilarity = squareform(dissimilarity_matrix)

# `linkage` computes the hierarchical clustering from a condensed distance matrix.
# The 'complete' method corresponds to complete linkage.
Z = linkage(condensed_dissimilarity, method='complete')

# Plot the dendrogram
plt.figure(figsize=(8, 6))
dendrogram(Z, labels=['Obs 1', 'Obs 2', 'Obs 3', 'Obs 4'])
plt.title('Dendrogram for Complete Linkage')
plt.ylabel('Dissimilarity')
plt.xlabel('Observations')
plt.grid(True)
plt.show()

# Print the linkage matrix to verify the fusion steps
print("Linkage Matrix (Z):")
print(Z)
```
