(e) Use the boxplot() method of college to produce side-by-side

(f) Create a new qualitative variable, called Elite, by binning the
Top10perc variable into two groups based on whether or not the
proportion of students coming from the top 10% of their high
school classes exceeds 50%.

```python
college['Elite']=pd.cut(college['Top10perc'],
                        [0,50,100],
                        labels=['No','Yes'])
```

Use the value_counts() method of college['Elite'] to see how
many elite universities there are. Finally, use the boxplot() method
again to produce side-by-side boxplots of Outstate versus Elite.

```python
print(college['Elite'].value_counts())
college.boxplot(column="Outstate",by="Elite",figsize=(8,6))

plt.title("Outstate vs Elite")
plt.suptitle("")
plt.xlabel("Elite(Yes/No)")
plt.ylabel("Out-of-State Tuition")
plt.show()
```

(g) Use the plot.hist() method of college to produce some histograms with differing numbers of bins for a few of the quantitative variables. The command plt.subplots(2, 2) may be useful: it will divide the plot window into four regions so that four
plots can be made simultaneously. By changing the arguments
you can divide the screen up in other combinations.

```python
fig, axes = plt.subplots(2,2,figsize=(12,10),sharey=False)

vars_to_plot = ["Apps","Accept","Enroll",'Outstate']

college[vars_to_plot[0]].plot.hist(bins=10, ax=axes[0,0],color='skyblue',edgecolor='black')
axes[0,0].set_title("Apps (10 bins)")

college[vars_to_plot[1]].plot.hist(bins=20, ax=axes[0,1],color='lightgreen',edgecolor='black')
axes[0,1].set_title("Accept (20 bins)")

college[vars_to_plot[2]].plot.hist(bins=30, ax=axes[1,0],color='salmon',edgecolor='black')
axes[1,0].set_title("Enroll (30 bins)")

college[vars_to_plot[3]].plot.hist(bins=40, ax=axes[1,1],color='orange',edgecolor='black')
axes[1,1].set_title("Outstate (40 bins)")
```
