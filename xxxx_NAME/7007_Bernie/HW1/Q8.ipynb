# A Summary of my observation 
Private institutions tend to charge higher out-of-state tuition compared to public ones, but they also show higher graduation rates on average. Schools with a greater percentage of students from the top 10% of their high school class often report higher instructional expenditure and alumni donation rates. In addition, institutions with more full-time undergraduates usually have larger room and board costs. These findings suggest that academic selectivity, financial resources, and student outcomes are closely connected.
