#question_b
Look at the data used in the notebook by creating and running a new cell with just the code college in it. You should notice that the first column is just the name of each university in a column named something like Unnamed: 0. We don’t really want pandas to treat this as data. However, it may be handy to have these names for later.

This has used the first column in the file as an index for the data frame. This means that pandas has given each row a name corresponding to the appropriate university. Now you should see that the first data column is Private. Note that the names of the colleges appear on the left of the table. #question_f
Create a new qualitative variable, called Elite,bybinning the Top10perc variable into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.

Use the value_counts() method of college['Elite'] to see how manyelite universities there are. Finally, use the boxplot() method again to produce side-by-side boxplots of Outstate versus Elite.

```python
data['Elite'] = pd.cut(data['Top10perc'], [0,50,100], labels=['No', 'Yes'])
data['Elite'].value_counts()

import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots

fig, ax = subplots(figsize=(8,8))
data.boxplot('Outstate', by='Elite', ax=ax)
plt.show()
```

#question_g
Use the plot.hist() method of college to produce some histograms with differing numbers of bins for a few of the quanti
tative variables. The command plt.subplots(2, 2) may be useful: it will divide the plot window into four regions so that four plots can be made simultaneously. By changing the arguments you can divide the screen up in other combinations.

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
#1
data['Apps'].hist(bins=10, ax=axes[0, 0], color='gray')
axes[0, 0].set_title('App bins=10')
#2
data['Apps'].hist(bins=20, ax=axes[0, 1], color='gray')
axes[0, 1].set_title('App bins=20')
#3
data['Apps'].hist(bins=50, ax=axes[1, 0], color='gray')
axes[1, 0].set_title('App bins=50')
#4
data['Apps'].hist(bins=100, ax=axes[1, 1], color='gray')
axes[1, 1].set_title('App bins=100')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

#1
data.boxplot('Outstate', by='Private', ax=axes[0, 0])
axes[0, 0].set_title('Outstate Tuition vs. Private')
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('Tuition')

#2
data.boxplot('Room.Board', by='Private', ax=axes[0, 1])
axes[0, 1].set_title('Room & Board vs. Private')
axes[0, 1].set_xlabel('')
axes[0, 1].set_ylabel('Cost')

#3
data.boxplot('Books', by='Private', ax=axes[1, 0])
axes[1, 0].set_title('Books vs. Private')
axes[1, 0].set_xlabel('')
axes[1, 0].set_ylabel('Cost')

#4
data.boxplot('Personal', by='Private', ax=axes[1, 1])
axes[1, 1].set_title('Personal Expenses vs. Private')
axes[1, 1].set_xlabel('')
axes[1, 1].set_ylabel('Cost')

plt.suptitle('Comparison of Public vs. Private University Costs', y=1.02, fontsize=16)
plt.tight_layout()

plt.show()
```
