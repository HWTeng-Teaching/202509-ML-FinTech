## (a) Import the package and read **Coolege.csv**.

```python
import pandas as pd
import matplotlib.pyplot as plt

college = pd.read_csv('/content/College.csv')
```

## (b) Rename "the name of each university" column name to College.

```python
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)
college3 = college3.set_index('College')
college = college3

display(college.head())
```

## (c) Use the describe() method of to produce a numerical summary of the variables in the data set.
We can see many datas about dataset such as mean, std, min...

```python
college.describe()
```

## (d) Use the pd.plotting.scatter_matrix() function to produce ascatterplot matrix of the frst columns [Top10perc, Apps, Enroll].

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

## (e) Use the boxplot() method of college to produce side-by-sideboxplots of Outstate versus Private.

```python
college.boxplot(column='Outstate', by='Private')
```

## (f) Create a new qualitative variable, called Elite, by binning theTop10perc variable into two groups based on whether or not theproportion of students coming from the top 10% of their high school classes exceeds 50%.

```python
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
display(college['Elite'].value_counts())

college.boxplot(column='Outstate', by='Elite')
```

## (g) Use the plot.hist() method of college to produce some histograms with difering numbers of bins for a few of the quantitative variables.

I choose 'Outstate', 'Enroll', 'Grad.Rate', and 'Expand' with 10 bins, where each represents cost, size, student success, and resources. This allows us to see the differences across the four aspects of universities in the chart.

```python
import matplotlib.pyplot as plt

plt.subplots(2, 2)

college['Outstate'].plot.hist(ax=plt.subplot(2, 2, 1), bins=10, title='Out-of-State Tuition')
college['Enroll'].plot.hist(ax=plt.subplot(2, 2, 2), bins=10, title='Enrollment')
college['Grad.Rate'].plot.hist(ax=plt.subplot(2, 2, 3), bins=10, title='Graduation Rate')
college['Expend'].plot.hist(ax=plt.subplot(2, 2, 4), bins=10, title='Instructional Expenditure per Student')
plt.tight_layout()
plt.show()
```

## (h) Other discovery


### I. Correlations of each varialbes
To explore relationships between variables and investigate correlations among different quantitative variables, calculate the correlation matrix for all quantitative variables and visualize their relationships using a heatmap.

```python
correlation_matrix = college.corr(numeric_only=True)
display(correlation_matrix)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Quantitative Variables')
plt.show()
```

### II. Analyze 'private' vs. 'public' universities
To analyze differences between private and public universities across several quantitative variables, generate box plots to compare their distributions.

```python
display(college.groupby('Private')[['Apps', 'Accept', 'Enroll', 'Outstate', 'Grad.Rate', 'Expend']].describe())
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
college.boxplot(column='Apps', by='Private', ax=axes[0, 0])
college.boxplot(column='Accept', by='Private', ax=axes[0, 1])
college.boxplot(column='Enroll', by='Private', ax=axes[0, 2])
college.boxplot(column='Outstate', by='Private', ax=axes[1, 0])
college.boxplot(column='Grad.Rate', by='Private', ax=axes[1, 1])
college.boxplot(column='Expend', by='Private', ax=axes[1, 2])
plt.tight_layout()
plt.show()
```
