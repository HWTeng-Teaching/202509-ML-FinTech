(f) Create a new qualitative variable, called Elite, by binning the Top10perc variable into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.Use the value_counts() method of college['Elite'] to see how many elite universities there are. (g) Use the plot.hist() method of college to produce some histograms with differing numbers of bins for a few of the quantitative variables. 
The command plt.subplots(2, 2) may be useful: it will divide the plot window into four regions so that four plots can be made simultaneously. 
By changing the arguments you can divide the screen up in other combinations.

```python
fig, axes = plt.subplots(2, 2) 
college['Outstate'].plot.hist(bins=100, ax=axes[0,0]) #bins把數值範圍切成幾個小區間，bins 越多，直方圖越精細；bins 越少，資料越粗略
axes[0, 0].set_title('Outstate')
college['Apps'].plot.hist(bins=20, ax=axes[0,1])
axes[0, 1].set_title('Apps')
college['Accept'].plot.hist(bins=15, ax=axes[1,0])
axes[1, 0].set_title('Accept')
college['Enroll'].plot.hist(bins=25, ax=axes[1,1])
axes[1, 1].set_title('Enroll')
plt.tight_layout() 
```

(h) Continue exploring the data, and provide a brief summary of what you discover.

```python
#相關係數熱圖
import seaborn as sns
sns.heatmap(college.corr(numeric_only=True), cmap="coolwarm", annot=False) 
plt.title("Correlation Heatmap of College Data")
```

1. Strong Correlations

* Top10perc and Top25perc are almost perfectly positively correlated (>0.9), since both measure students’ academic background.

* Apps (applications), Accept (admissions), and Enroll (enrollments) are highly positively correlated, as more applications generally lead to more admissions and more enrollments.

* Expend (expenditure) and Outstate (out-of-state tuition) are strongly positively correlated — schools that spend more tend to charge higher tuition fees.

2. Negative Correlations

* S.F.Ratio (student-faculty ratio) and Expend are negatively correlated, indicating that schools with higher expenditures typically have lower student-faculty ratios, possibly due to richer faculty resources.
