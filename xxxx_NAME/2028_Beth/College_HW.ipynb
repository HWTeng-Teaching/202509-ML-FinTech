```python
df['Elite']=pd.cut(df['Top10perc'],bins=[0,50,100],labels=['No','Yes'])
print(df['Elite'].value_counts())
```

```python
df.boxplot(column='Outstate',by='Elite',figsize=(6,6))
plt.ylabel('Outstate Tuition')
plt.show()
```

```python

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

df['Apps'].plot.hist(bins=20, ax=axes[0,0], title='Applications')
df['Accept'].plot.hist(bins=20, ax=axes[0,1],title='Acceptances')
df['Enroll'].plot.hist(bins=20, ax=axes[1,0],title='Enrollments')
df['Outstate'].plot.hist(bins=20, ax=axes[1,1],title='Outstate Tuition')
plt.tight_layout()
plt.show()
```
