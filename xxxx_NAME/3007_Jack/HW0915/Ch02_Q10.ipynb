```python
fig, axes = plt.subplots(1, 3)
Boston['crim'].plot.hist(bins=30, ax=axes[0], title='crim')
Boston['tax'].plot.hist(bins=30, ax=axes[1], title='tax')
Boston['ptratio'].plot.hist(bins=30, ax=axes[2], title='ptratio')
plt.tight_layout()
plt.show()
```

### crim
 - Most suburbs have crime rates below 10
 - A few suburbs show extremely high crime rates
 - The distribution is highly right-skewed with significant outliers
 - 少數郊區顯示極高的犯罪率，存在明顯異常值
### tax
 - Majority of suburbs have tax rates between 200-400
 - Some suburbs have particularly high tax rates
 - Bimodal distribution suggests two distinct groups of suburbs
 - 部分郊區稅率特別高，存在兩個不同的郊區群組
### ptratio
 - Most ratios cluster between 14-21
 - Some suburbs have very high ratios
