(g) Use the plot.hist() method of college to produce some histograms 直方圖 with differing numbers of bins for a few of the quanti
tative variables. The command plt.subplots(2, 2) may be useful: it will divide the plot window into four regions so that four
plots can be made simultaneously. By changing the arguments
you can divide the screen up in other combinations.

````python
# 建立 2x2 畫布
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 直方圖 1：Top10perc，切 10 個 bins
college['Top10perc'].plot.hist(bins=10, ax=axes[0, 0])
axes[0, 0].set_title('Top10perc')

# 直方圖 2：Apps，切 20 個 bins
college['Apps'].plot.hist(bins=20, ax=axes[0, 1])
axes[0, 1].set_title('Apps')

# 直方圖 3：Accept，切 30 個 bins
college['Accept'].plot.hist(bins=30, ax=axes[1, 0])
axes[1, 0].set_title('Accept')

# 直方圖 4：Outstate，切 50 個 bins
college['Outstate'].plot.hist(bins=50, ax=axes[1, 1])
axes[1, 1].set_title('Outstate')

plt.tight_layout()  # 自動調整間距避免重疊
plt.show()
````
