(g) Use the plot.hist() method of college to produce some his- tograms with differing numbers of bins for a few of the quanti- tative variables. The command plt.subplots(2, 2) may be use- ful: it will divide the plot window into four regions so that four plots can be made simultaneously. By changing the arguments you can divide the screen up in other combinations.

```python
import matplotlib.pyplot as plt

# 選擇幾個定量變數，例如 'F.Undergrad', 'Outstate', 'Room.Board', 'Grad.Rate'
vars = ['F.Undergrad', 'Outstate', 'Room.Board', 'Grad.Rate']

# 建立4個子圖
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 在每個子圖繪製直方圖，並設定不同的 bin 數量
axes[0, 0].hist(college['F.Undergrad'], bins=20)
axes[0, 0].set_title('F.Undergrad - 20 bins')

axes[0, 1].hist(college['Outstate'], bins=30)
axes[0, 1].set_title('Outstate - 30 bins')

axes[1, 0].hist(college['Room.Board'], bins=15)
axes[1, 0].set_title('Room.Board - 15 bins')

axes[1, 1].hist(college['Grad.Rate'], bins=25)
axes[1, 1].set_title('Grad.Rate - 25 bins')

# 顯示圖形
plt.tight_layout()
plt.show()
```
