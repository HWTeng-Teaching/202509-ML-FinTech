```python
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1. Define the standard points ---
P1 = np.array([2, 2])
P2 = np.array([4, 0])

# --- Step 2. Define the new orthonormal basis (45° rotation) ---
u1 = np.array([1, 1]) / np.sqrt(2)
u2 = np.array([-1, 1]) / np.sqrt(2)

# --- Step 3. Compute coordinates in the new basis using inner product ---
def coordinates_in_new_basis(P, u1, u2):
    c1 = np.dot(u1, P)   # <u1, P>
    c2 = np.dot(u2, P)   # <u2, P>
    return np.array([c1, c2])

cP1 = coordinates_in_new_basis(P1, u1, u2)
cP2 = coordinates_in_new_basis(P2, u1, u2)

print("Coordinates of P1 in new basis:", cP1)
print("Coordinates of P2 in new basis:", cP2)

# --- Step 4. Visualization ---
fig, ax = plt.subplots(figsize=(6,6))

# Plot standard axes
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)

# Plot new basis vectors
ax.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u1')
ax.quiver(0, 0, u2[0], u2[1], angles='xy', scale_units='xy', scale=1, color='green', label='u2')

# Plot points
ax.plot(P1[0], P1[1], 'ro', label='P1 (2,2)')
ax.plot(P2[0], P2[1], 'mo', label='P2 (4,0)')

# Annotate projections
ax.text(P1[0]+0.1, P1[1], f"P1\ncoords={cP1.round(2)}", color='red')
ax.text(P2[0]+0.1, P2[1], f"P2\ncoords={cP2.round(2)}", color='purple')

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_aspect('equal')
ax.legend()
ax.set_title("Basis Transformation Example (Standard vs Orthonormal Basis)")
plt.show()
```
